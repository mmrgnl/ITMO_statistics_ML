## 1.1 Вероятностное пространство и формула Байеса


### Задача 1.1.1. Исследование независимости событий

Определить (с обоснованием), зависимы или независимы следующие события:
- **(a)** Несовместные события
- **(b)** События, образующие $\sigma$-алгебру $\Sigma$ в пространстве $(\Omega, \Sigma, \mathbb{P})$
- **(c)** События, имеющие одинаковую вероятность


#### Решение

**Определение независимости:** События $A$ и $B$ называются независимыми, если выполняется условие:
$$\mathbb{P}(A \cap B) = \mathbb{P}(A) \cdot \mathbb{P}(B)$$

**Эквивалентное определение:** События независимы, если $\mathbb{P}(A \mid B) = \mathbb{P}(A)$ при $\mathbb{P}(B) > 0$.

---

**(a) Несовместные события**

Если события $A$ и $B$ несовместны, то $A \cap B = \varnothing$, откуда $\mathbb{P}(A \cap B) = 0$.

Для независимости должно выполняться: $\mathbb{P}(A) \cdot \mathbb{P}(B) = 0$.

Это возможно только в вырожденных случаях:
- $\mathbb{P}(A) = 0$ (событие невозможно)
- $\mathbb{P}(B) = 0$ (событие невозможно)

**Вывод:** Несовместные события с ненулевыми вероятностями **зависимы**.

**Обоснование:** Если $A$ и $B$ несовместны и $\mathbb{P}(A) > 0$, $\mathbb{P}(B) > 0$, то условная вероятность $\mathbb{P}(A \mid B) = 0$, в то время как безусловная $\mathbb{P}(A) > 0$. Следовательно, события зависимы.

---

**(b) События из $\sigma$-алгебры**

$\sigma$-алгебра $\Sigma$ — это система подмножеств пространства $\Omega$, которая:
- содержит $\varnothing$ и $\Omega$
- замкнута относительно дополнений и счётных объединений

**Вывод:** События из $\sigma$-алгебры могут быть как зависимыми, так и независимыми — это зависит от конкретной меры $\mathbb{P}$.

**Примеры:**

*Зависимые события:*
- В пространстве $\Omega = \{1,2,3,4\}$ с равномерной мерой рассмотрим $\Sigma = \{\varnothing, \Omega, \{1,2\}, \{3,4\}\}$
- События $A = \{1,2\}$ и $B = \{3,4\}$ принадлежат $\Sigma$
- $\mathbb{P}(A) = \mathbb{P}(B) = 1/2$, но $\mathbb{P}(A \cap B) = 0 \neq 1/4$ → события зависимы

*Независимые события:*
- При бросании двух монет: $\Omega = \{ГГ, ГР, РГ, РР\}$ с равномерной вероятностью
- События $A = \{\text{герб на первой монете}\} = \{ГГ, ГР\}$ и $B = \{\text{герб на второй монете}\} = \{ГГ, РГ\}$
- $\mathbb{P}(A) = \mathbb{P}(B) = 1/2$, $\mathbb{P}(A \cap B) = \mathbb{P}(\{ГГ\}) = 1/4 = \mathbb{P}(A)\mathbb{P}(B)$ → события независимы

---

**(c) События одинаковой вероятности**

**Вывод:** Одинаковая вероятность не гарантирует независимость.

**Примеры:**

*Зависимые равновероятные события:*
- При бросании одной честной кости: $A = \{1,2,3\}$, $B = \{2,3,4\}$
- $\mathbb{P}(A) = \mathbb{P}(B) = 1/2$
- Но $\mathbb{P}(A \cap B) = \mathbb{P}(\{2,3\}) = 1/3 \neq 1/4$ → события зависимы

*Независимые равновероятные события:*
- При бросании двух монет: $A = \{\text{герб на первой}\}$, $B = \{\text{герб на второй}\}$
- $\mathbb{P}(A) = \mathbb{P}(B) = 1/2$
- $\mathbb{P}(A \cap B) = 1/4 = \mathbb{P}(A)\mathbb{P}(B)$ → события независимы


### Задача 1.1.2. Операции над событиями при бросании двух монет

Опыт заключается в независимом подбрасывании двух симметричных монет. Рассматриваются следующие события:

- $A$ — появление герба на первой монете
- $B$ — появление решки на первой монете  
- $C$ — появление герба на второй монете
- $D$ — появление решки на второй монете
- $E$ — появление хотя бы одного герба
- $F$ — появление хотя бы одной решки
- $G$ — появление одного герба и одной решки
- $H$ — непоявление ни одного герба
- $K$ — появление двух гербов

Определить, каким событиям этого списка равносильны следующие события:
- (a) $A + C = ?$
- (b) $AC = ?$
- (c) $EF = ?$
- (d) $G + E = ?$
- (e) $GE = ?$
- (f) $BD = ?$
- (g) $E + K = ?$


#### Решение

Пространство элементарных исходов: $\Omega = \{ГГ, ГР, РГ, РР\}$, где первая буква — результат первой монеты, вторая — второй.

Определим множества исходов для каждого события:

| Событие | Описание | Множество исходов |
|---------|----------|-------------------|
| $A$ | Герб на первой монете | $\{ГГ, ГР\}$ |
| $B$ | Решка на первой монете | $\{РГ, РР\}$ |
| $C$ | Герб на второй монете | $\{ГГ, РГ\}$ |
| $D$ | Решка на второй монете | $\{ГР, РР\}$ |
| $E$ | Хотя бы один герб | $\{ГГ, ГР, РГ\}$ |
| $F$ | Хотя бы одна решка | $\{ГР, РГ, РР\}$ |
| $G$ | Ровно один герб и одна решка | $\{ГР, РГ\}$ |
| $H$ | Ни одного герба | $\{РР\}$ |
| $K$ | Два герба | $\{ГГ\}$ |

Теперь найдём эквивалентности:

**(a)** $A + C = A \cup C = \{ГГ, ГР\} \cup \{ГГ, РГ\} = \{ГГ, ГР, РГ\} = E$  
**Ответ:** $A + C = E$

**(b)** $AC = A \cap C = \{ГГ, ГР\} \cap \{ГГ, РГ\} = \{ГГ\} = K$  
**Ответ:** $AC = K$

**(c)** $EF = E \cap F = \{ГГ, ГР, РГ\} \cap \{ГР, РГ, РР\} = \{ГР, РГ\} = G$  
**Ответ:** $EF = G$

**(d)** $G + E = G \cup E = \{ГР, РГ\} \cup \{ГГ, ГР, РГ\} = \{ГГ, ГР, РГ\} = E$  
**Ответ:** $G + E = E$

**(e)** $GE = G \cap E = \{ГР, РГ\} \cap \{ГГ, ГР, РГ\} = \{ГР, РГ\} = G$  
**Ответ:** $GE = G$

**(f)** $BD = B \cap D = \{РГ, РР\} \cap \{ГР, РР\} = \{РР\} = H$  
**Ответ:** $BD = H$

**(g)** $E + K = E \cup K = \{ГГ, ГР, РГ\} \cup \{ГГ\} = \{ГГ, ГР, РГ\} = E$  
**Ответ:** $E + K = E$

### Задача 1.1.4. Вероятность ожидания пароходов у причала

Два парохода должны подойти к одному и тому же причалу независимо друг от друга и равновозможно в течение суток. Определить вероятность того, что одному из них придётся ожидать освобождения причала, если время стоянки первого парохода — 1 час, а второго — 2 часа.


#### Решение

Обозначим:
- $x \in [0, 24]$ — время прибытия первого парохода (в часах)
- $y \in [0, 24]$ — время прибытия второго парохода (в часах)

Ожидание возникает в следующих случаях:

1. **Второй пароход ждёт первого:** если $x < y$ и $y - x < 1$ (второй пришёл менее чем через час после первого)
2. **Первый пароход ждёт второго:** если $y < x$ и $x - y < 2$ (первый пришёл менее чем через 2 часа после второго)

Объединяя условия: ожидание происходит, если $-2 < x - y < 1$.

Геометрически это полоса между прямыми $y = x - 2$ и $y = x + 1$ внутри квадрата $[0, 24] \times [0, 24]$.

Площадь области ожидания можно вычислить как площадь квадрата минус площади двух прямоугольных треугольников, где ожидания нет:

- **Верхний треугольник** (над $y = x + 1$): катеты по 23, площадь $S_1 = \frac{23 \times 23}{2} = 264.5$
- **Нижний треугольник** (под $y = x - 2$): катеты по 22, площадь $S_2 = \frac{22 \times 22}{2} = 242$

Площадь квадрата: $S_{\text{общ}} = 24 \times 24 = 576$

Площадь области ожидания: $S_{\text{ожид}} = 576 - 264.5 - 242 = 69.5$

Вероятность:
$$\mathbb{P}(\text{ожидание}) = \frac{69.5}{576} = \frac{139}{1152}$$

**Ответ:** $\boxed{\dfrac{139}{1152} \approx 0.1207}$


```python
# Численная проверка методом Монте-Карло
import numpy as np

def monte_carlo_probability(n_samples=1000000):
    """Оценка вероятности ожидания методом Монте-Карло"""
    np.random.seed(42)
    x = np.random.uniform(0, 24, n_samples)
    y = np.random.uniform(0, 24, n_samples)
    
    # Условие ожидания: -2 < x - y < 1
    waiting = np.sum((-2 < x - y) & (x - y < 1))
    return waiting / n_samples

# Вычисляем вероятность
prob_mc = monte_carlo_probability()
prob_exact = 139 / 1152
print('Точное значение:', prob_exact)
print('Монте-Карло (1M):', prob_mc)
print(f"Относительная ошибка: {abs(prob_mc - prob_exact) / prob_exact * 100:.4f}%")
```

### Задача 1.1.5. Вероятность поражения самолёта

Самолёт, по которому ведётся стрельба, состоит из трёх различных по уязвимости частей:
- **(a)** Кабина летчика и двигатель — достаточно одного попадания
- **(b)** Топливные баки — нужно два попадания
- **(c)** Планер — нужно три попадания

При попадании в самолёт одного снаряда, снаряд с вероятностью $p_1$ попадает в первую часть, с вероятностью $p_2$ — во вторую, с вероятностью $p_3$ — в третью ($p_1 + p_2 + p_3 = 1$). Попавшие снаряды распределяются по частям независимо друг от друга.

Известно, что в самолёт попало $m$ снарядов. Найти условную вероятность $\mathbb{P}(A \mid m)$ события $A$ — «Самолёт поражён» — при $m = 1, 2, 3, 4$.


#### Решение

Обозначим через $X_1$, $X_2$, $X_3$ количество попаданий в первую, вторую и третью части соответственно. При фиксированном $m$ вектор $(X_1, X_2, X_3)$ имеет мультиномиальное распределение с параметрами $(m, p_1, p_2, p_3)$.

Самолёт поражён (событие $A$), если выполнено хотя бы одно из условий:
- $X_1 \geq 1$ (хотя бы одно попадание в кабину/двигатель)
- $X_2 \geq 2$ (не менее двух попаданий в топливные баки)
- $X_3 \geq 3$ (не менее трёх попаданий в планер)

Удобнее вычислять вероятность дополнительного события $\bar{A}$ (самолёт не поражён) и затем использовать $\mathbb{P}(A \mid m) = 1 - \mathbb{P}(\bar{A} \mid m)$.

Самолёт не поражён, если:
- $X_1 = 0$ (ни одного попадания в первую часть)
- $X_2 \leq 1$ (не более одного попадания во вторую часть)
- $X_3 \leq 2$ (не более двух попаданий в третью часть)

---

**Случай $m = 1$:**

Единственный способ избежать поражения — промахнуться по первой части:
$$\mathbb{P}(\bar{A} \mid 1) = p_2 + p_3 = 1 - p_1$$

$$\boxed{\mathbb{P}(A \mid 1) = p_1}$$

---

**Случай $m = 2$:**

Самолёт цел только в случаях:
- $(X_1, X_2, X_3) = (0, 0, 2)$: вероятность $p_3^2$
- $(X_1, X_2, X_3) = (0, 1, 1)$: вероятность $2p_2 p_3$ (два способа распределения)

$$\mathbb{P}(\bar{A} \mid 2) = p_3^2 + 2p_2 p_3$$

$$\boxed{\mathbb{P}(A \mid 2) = 1 - p_3^2 - 2p_2 p_3}$$

---

**Случай $m = 3$:**

Единственная безопасная конфигурация:
- $(X_1, X_2, X_3) = (0, 1, 2)$: вероятность $3p_2 p_3^2$ (три способа выбора, какой снаряд попал во вторую часть)

$$\mathbb{P}(\bar{A} \mid 3) = 3p_2 p_3^2$$

$$\boxed{\mathbb{P}(A \mid 3) = 1 - 3p_2 p_3^2}$$

---

**Случай $m = 4$:**

При $X_1 = 0$ имеем $X_2 + X_3 = 4$. Но $X_2 \leq 1$ и $X_3 \leq 2$, откуда $X_2 + X_3 \leq 3 < 4$ — противоречие.

Следовательно, безопасных конфигураций нет:
$$\boxed{\mathbb{P}(A \mid 4) = 1}$$

---

**Итоговые ответы:**

$$
\begin{aligned}
\mathbb{P}(A \mid 1) &= p_1 \\
\mathbb{P}(A \mid 2) &= 1 - p_3^2 - 2p_2 p_3 \\
\mathbb{P}(A \mid 3) &= 1 - 3p_2 p_3^2 \\
\mathbb{P}(A \mid 4) &= 1
\end{aligned}
$$


## 1.2 Случайный вектор и числовые характеристики


### Задача 1.2.1. Проверка функции плотности распределения

Пусть
$$f_\xi(x, y) = \frac{e^{-2|y|}}{\pi(1+x^2)}$$

Является ли данная функция плотностью распределения случайного вектора?


#### Решение

Для того чтобы функция $f(x, y)$ была плотностью распределения случайного вектора на $\mathbb{R}^2$, необходимо и достаточно выполнение двух условий:

1. **Неотрицательность:** $f(x, y) \geq 0$ для всех $(x, y) \in \mathbb{R}^2$
2. **Нормировка:** $\iint_{\mathbb{R}^2} f(x, y) \, dx \, dy = 1$

---

**Проверка неотрицательности:**

Все компоненты функции положительны:
- $e^{-2|y|} > 0$ для всех $y \in \mathbb{R}$
- $\pi > 0$
- $1 + x^2 > 0$ для всех $x \in \mathbb{R}$

Следовательно, $f_\xi(x, y) > 0$ для всех $(x, y) \in \mathbb{R}^2$. ✓

---

**Проверка нормировки:**

Интеграл можно разложить на произведение двух одномерных интегралов:

$$\iint_{\mathbb{R}^2} \frac{e^{-2|y|}}{\pi(1+x^2)} \, dx \, dy = \frac{1}{\pi} \int_{-\infty}^{\infty} \frac{dx}{1+x^2} \cdot \int_{-\infty}^{\infty} e^{-2|y|} \, dy$$

**Первый интеграл:**
$$\int_{-\infty}^{\infty} \frac{dx}{1+x^2} = \arctan x \Big|_{-\infty}^{\infty} = \frac{\pi}{2} - \left(-\frac{\pi}{2}\right) = \pi$$

**Второй интеграл:**
$$\int_{-\infty}^{\infty} e^{-2|y|} \, dy = 2 \int_0^{\infty} e^{-2y} \, dy = 2 \cdot \left[-\frac{1}{2}e^{-2y}\right]_0^{\infty} = 2 \cdot \frac{1}{2} = 1$$

**Итого:**
$$\iint_{\mathbb{R}^2} f_\xi(x, y) \, dx \, dy = \frac{1}{\pi} \cdot \pi \cdot 1 = 1$$ 

---

**Вывод:** Функция удовлетворяет обоим условиям и является плотностью распределения случайного вектора.

**Замечание:** Данная плотность представляет собой произведение плотностей:
- Распределения Коши по $x$: $f_X(x) = \frac{1}{\pi(1+x^2)}$
- Симметричного экспоненциального распределения по $y$: $f_Y(y) = e^{-2|y|}$

Это означает, что компоненты вектора независимы.


```python
# Проверка нормировки с помощью символьных вычислений
import sympy as sp

x, y = sp.symbols('x y', real=True)
f = sp.exp(-2*sp.Abs(y)) / (sp.pi * (1 + x**2))

# Вычисляем интеграл
integral_x = sp.integrate(1/(1 + x**2), (x, -sp.oo, sp.oo))
integral_y = sp.integrate(sp.exp(-2*sp.Abs(y)), (y, -sp.oo, sp.oo))

total_integral = (1/sp.pi) * integral_x * integral_y

print("Интеграл от -inf до +inf dx/(1+x^2) =", integral_x)
print("Интеграл от -inf до +inf e^(-2|y|) dy =", integral_y)
print(f"\nОбщий интеграл: {sp.simplify(total_integral)}")
print(f"Проверка: {total_integral == 1}")
```

### Задача 1.2.2. Совместное распределение дискретных случайных величин

Совместное распределение случайных величин $\xi$ и $\eta$ задано следующей таблицей:

| $\xi \backslash \eta$ | $-1$   | $0$    | $1$    |
|-----------------------|--------|--------|--------|
| $-1$                  | $1/8$  | $1/12$ | $7/24$ |
| $1$                   | $1/3$  | $1/6$  | $0$    |

**(a)** Найти маргинальные распределения $\xi$ и $\eta$  
**(b)** Вычислить математическое ожидание, ковариационную и корреляционную матрицы вектора $(\xi, \eta)$  
**(c)** Исследовать $\xi$ и $\eta$ на независимость и некоррелированность


#### Решение

**(a) Маргинальные распределения**

**Распределение $\xi$:**
$$\mathbb{P}(\xi = -1) = \frac{1}{8} + \frac{1}{12} + \frac{7}{24} = \frac{3}{24} + \frac{2}{24} + \frac{7}{24} = \frac{12}{24} = \frac{1}{2}$$

$$\mathbb{P}(\xi = 1) = \frac{1}{3} + \frac{1}{6} + 0 = \frac{2}{6} + \frac{1}{6} = \frac{3}{6} = \frac{1}{2}$$

**Распределение $\eta$:**
$$\mathbb{P}(\eta = -1) = \frac{1}{8} + \frac{1}{3} = \frac{3}{24} + \frac{8}{24} = \frac{11}{24}$$

$$\mathbb{P}(\eta = 0) = \frac{1}{12} + \frac{1}{6} = \frac{1}{12} + \frac{2}{12} = \frac{3}{12} = \frac{1}{4}$$

$$\mathbb{P}(\eta = 1) = \frac{7}{24} + 0 = \frac{7}{24}$$

Проверка нормировки: $\frac{11}{24} + \frac{1}{4} + \frac{7}{24} = \frac{11+6+7}{24} = 1$ ✓

---

**(b) Числовые характеристики**

**Математические ожидания:**
$$E[\xi] = (-1) \cdot \frac{1}{2} + 1 \cdot \frac{1}{2} = 0$$

$$E[\eta] = (-1) \cdot \frac{11}{24} + 0 \cdot \frac{1}{4} + 1 \cdot \frac{7}{24} = -\frac{11}{24} + \frac{7}{24} = -\frac{4}{24} = -\frac{1}{6}$$

**Дисперсии:**
$$E[\xi^2] = (-1)^2 \cdot \frac{1}{2} + 1^2 \cdot \frac{1}{2} = 1$$
$$D[\xi] = E[\xi^2] - (E[\xi])^2 = 1 - 0 = 1$$

$$E[\eta^2] = (-1)^2 \cdot \frac{11}{24} + 0^2 \cdot \frac{1}{4} + 1^2 \cdot \frac{7}{24} = \frac{11}{24} + \frac{7}{24} = \frac{18}{24} = \frac{3}{4}$$
$$D[\eta] = E[\eta^2] - (E[\eta])^2 = \frac{3}{4} - \left(-\frac{1}{6}\right)^2 = \frac{3}{4} - \frac{1}{36} = \frac{27}{36} - \frac{1}{36} = \frac{26}{36} = \frac{13}{18}$$

**Ковариация:**
$$E[\xi\eta] = (-1)(-1) \cdot \frac{1}{8} + (-1) \cdot 0 \cdot \frac{1}{12} + (-1) \cdot 1 \cdot \frac{7}{24} + 1 \cdot (-1) \cdot \frac{1}{3} + 1 \cdot 0 \cdot \frac{1}{6} + 1 \cdot 1 \cdot 0$$
$$= \frac{1}{8} - \frac{7}{24} - \frac{1}{3} = \frac{3}{24} - \frac{7}{24} - \frac{8}{24} = -\frac{12}{24} = -\frac{1}{2}$$

$$\operatorname{cov}(\xi, \eta) = E[\xi\eta] - E[\xi] \cdot E[\eta] = -\frac{1}{2} - 0 \cdot \left(-\frac{1}{6}\right) = -\frac{1}{2}$$

**Ковариационная матрица:**
$$K = \begin{pmatrix} D[\xi] & \operatorname{cov}(\xi, \eta) \\ \operatorname{cov}(\xi, \eta) & D[\eta] \end{pmatrix} = \begin{pmatrix} 1 & -1/2 \\ -1/2 & 13/18 \end{pmatrix}$$

**Корреляция:**
$$\rho = \frac{\operatorname{cov}(\xi, \eta)}{\sqrt{D[\xi] \cdot D[\eta]}} = \frac{-1/2}{\sqrt{1 \cdot 13/18}} = -\frac{1}{2} \cdot \sqrt{\frac{18}{13}} = -\frac{3}{\sqrt{26}}$$

**Корреляционная матрица:**
$$R = \begin{pmatrix} 1 & \rho \\ \rho & 1 \end{pmatrix} = \begin{pmatrix} 1 & -3/\sqrt{26} \\ -3/\sqrt{26} & 1 \end{pmatrix}$$

---

**(c) Исследование на независимость и некоррелированность**

**Некоррелированность:**
Поскольку $\operatorname{cov}(\xi, \eta) = -1/2 \neq 0$, случайные величины **коррелированы**.

**Независимость:**
Проверим условие независимости: $\mathbb{P}(\xi = x, \eta = y) = \mathbb{P}(\xi = x) \cdot \mathbb{P}(\eta = y)$ для всех $x, y$.

Возьмём $x = 1$, $y = 1$:
- $\mathbb{P}(\xi = 1, \eta = 1) = 0$ (из таблицы)
- $\mathbb{P}(\xi = 1) \cdot \mathbb{P}(\eta = 1) = \frac{1}{2} \cdot \frac{7}{24} = \frac{7}{48} \neq 0$

Условие независимости не выполняется → случайные величины **зависимы**.

**Вывод:** $\xi$ и $\eta$ являются **зависимыми и коррелированными** случайными величинами.


### Задача 1.2.3. Два тетраэдра: сумма и сравнение

Пусть имеются два одинаковых тетраэдра с числами $1, 2, 3, 4$ на гранях. Подкидываем оба и смотрим на выпавшие числа $\xi_1$ и $\xi_2$. Зададим следующие случайные величины:

$$\phi_1 = \xi_1 + \xi_2, \quad \phi_2 = \begin{cases} 1, & \xi_1 \leq \xi_2 \\ 0, & \text{иначе} \end{cases}$$

### Задача 1.2.4. Тригонометрические функции равномерной случайной величины

Пусть $\xi \sim U[-\pi, \pi]$ и $\eta_1 = \cos \xi$, $\eta_2 = \sin \xi$.

**(a)** Вычислить математическое ожидание, ковариационную и корреляционную матрицы вектора $(\eta_1, \eta_2)$  
**(b)** Исследовать $\eta_1$ и $\eta_2$ на независимость и некоррелированность


#### Решение

Плотность распределения $\xi$:
$$f_\xi(x) = \begin{cases} \frac{1}{2\pi}, & x \in [-\pi, \pi] \\ 0, & \text{иначе} \end{cases}$$

**(a) Числовые характеристики**

**Математические ожидания:**

$$E[\eta_1] = E[\cos \xi] = \int_{-\pi}^{\pi} \cos x \cdot \frac{1}{2\pi} \, dx = \frac{1}{2\pi} \left[\sin x\right]_{-\pi}^{\pi} = 0$$

$$E[\eta_2] = E[\sin \xi] = \int_{-\pi}^{\pi} \sin x \cdot \frac{1}{2\pi} \, dx = \frac{1}{2\pi} \left[-\cos x\right]_{-\pi}^{\pi} = 0$$

**Дисперсии:**

$$E[\eta_1^2] = E[\cos^2 \xi] = \int_{-\pi}^{\pi} \cos^2 x \cdot \frac{1}{2\pi} \, dx = \frac{1}{2\pi} \int_{-\pi}^{\pi} \frac{1 + \cos 2x}{2} \, dx$$
$$= \frac{1}{4\pi} \left[x + \frac{\sin 2x}{2}\right]_{-\pi}^{\pi} = \frac{1}{4\pi} \cdot 2\pi = \frac{1}{2}$$

$$D[\eta_1] = E[\eta_1^2] - (E[\eta_1])^2 = \frac{1}{2} - 0 = \frac{1}{2}$$

Аналогично:
$$E[\eta_2^2] = E[\sin^2 \xi] = \int_{-\pi}^{\pi} \sin^2 x \cdot \frac{1}{2\pi} \, dx = \frac{1}{2\pi} \int_{-\pi}^{\pi} \frac{1 - \cos 2x}{2} \, dx = \frac{1}{2}$$

$$D[\eta_2] = \frac{1}{2}$$

**Ковариация:**

$$E[\eta_1 \eta_2] = E[\cos \xi \sin \xi] = \int_{-\pi}^{\pi} \cos x \sin x \cdot \frac{1}{2\pi} \, dx$$
$$= \frac{1}{2\pi} \int_{-\pi}^{\pi} \frac{\sin 2x}{2} \, dx = \frac{1}{4\pi} \left[-\frac{\cos 2x}{2}\right]_{-\pi}^{\pi} = 0$$

$$\operatorname{cov}(\eta_1, \eta_2) = E[\eta_1 \eta_2] - E[\eta_1] \cdot E[\eta_2] = 0 - 0 = 0$$

**Ковариационная матрица:**
$$K = \begin{pmatrix} D[\eta_1] & \operatorname{cov}(\eta_1, \eta_2) \\ \operatorname{cov}(\eta_1, \eta_2) & D[\eta_2] \end{pmatrix} = \begin{pmatrix} 1/2 & 0 \\ 0 & 1/2 \end{pmatrix}$$

**Корреляционная матрица:**
Поскольку ковариация равна нулю, корреляция также равна нулю:
$$R = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}$$

---

**(b) Исследование на независимость и некоррелированность**

**Некоррелированность:**
Поскольку $\operatorname{cov}(\eta_1, \eta_2) = 0$, случайные величины **некоррелированы**.

**Независимость:**
Несмотря на нулевую ковариацию, $\eta_1$ и $\eta_2$ **зависимы**.

**Доказательство зависимости:**

Заметим, что для любого значения $\xi$ выполняется тождество:
$$\eta_1^2 + \eta_2^2 = \cos^2 \xi + \sin^2 \xi = 1$$

Это означает, что вектор $(\eta_1, \eta_2)$ всегда лежит на единичной окружности — между компонентами существует функциональная связь.

**Контрпример для независимости:**

Рассмотрим события:
- $A = \{\eta_1 > \sqrt{3}/2\}$ (вероятность положительна)
- $B = \{\eta_2 > \sqrt{3}/2\}$ (вероятность положительна)

Если бы $\eta_1$ и $\eta_2$ были независимы, то $\mathbb{P}(A \cap B) = \mathbb{P}(A) \cdot \mathbb{P}(B) > 0$.

Но если $\eta_1 > \sqrt{3}/2$ и $\eta_2 > \sqrt{3}/2$, то:
$$\eta_1^2 + \eta_2^2 > \frac{3}{4} + \frac{3}{4} = \frac{3}{2} > 1$$

Это противоречит тому, что $\eta_1^2 + \eta_2^2 = 1$ с вероятностью 1.

Следовательно, $\mathbb{P}(A \cap B) = 0 \neq \mathbb{P}(A) \cdot \mathbb{P}(B)$, что доказывает зависимость.

**Вывод:** $\eta_1$ и $\eta_2$ являются **некоррелированными, но зависимыми** случайными величинами.

**Замечание:** Это классический пример того, что некоррелированность не влечёт независимость. Нулевая ковариация означает лишь отсутствие линейной зависимости, но не исключает нелинейной функциональной связи.


### Задача 1.2.5. Плотность распределения суммы независимых случайных величин

Найти плотность распределения суммы двух независимых случайных величин $\xi$ и $\eta$, если $\xi \sim \operatorname{Exp}(2)$ и $\eta \sim U[0,1]$.


#### Решение

Плотности распределения исходных случайных величин:

$$f_\xi(x) = \begin{cases} 2e^{-2x}, & x \geq 0 \\ 0, & x < 0 \end{cases}$$

$$f_\eta(y) = \begin{cases} 1, & y \in [0, 1] \\ 0, & \text{иначе} \end{cases}$$

Пусть $\zeta = \xi + \eta$. Поскольку $\xi$ и $\eta$ независимы, плотность распределения суммы находится по формуле свёртки:

$$f_\zeta(z) = \int_{-\infty}^{\infty} f_\xi(x) f_\eta(z-x) \, dx$$

Заметим, что:
- $f_\xi(x) \neq 0$ только при $x \geq 0$
- $f_\eta(z-x) \neq 0$ только при $0 \leq z-x \leq 1$, т.е. $z-1 \leq x \leq z$

Следовательно, область интегрирования определяется пересечением: $x \in [\max(0, z-1), z]$ при условии $z \geq 0$.

Рассмотрим случаи:

---

**Случай 1: $z < 0$**

Область интегрирования пуста → $f_\zeta(z) = 0$

---

**Случай 2: $0 \leq z < 1$**

В этом случае $z-1 < 0$, поэтому область интегрирования: $x \in [0, z]$

$$f_\zeta(z) = \int_0^z 2e^{-2x} \cdot 1 \, dx = 2 \left[-\frac{1}{2}e^{-2x}\right]_0^z = 1 - e^{-2z}$$

---

**Случай 3: $z \geq 1$**

В этом случае область интегрирования: $x \in [z-1, z]$

$$f_\zeta(z) = \int_{z-1}^z 2e^{-2x} \cdot 1 \, dx = \left[-e^{-2x}\right]_{z-1}^z = -e^{-2z} + e^{-2(z-1)} = e^{-2(z-1)} - e^{-2z}$$

$$= e^{-2z}(e^2 - 1) = (e^2 - 1)e^{-2z}$$

---

**Итоговая формула:**

$$\boxed{f_\zeta(z) = \begin{cases} 0, & z < 0 \\[6pt] 1 - e^{-2z}, & 0 \leq z < 1 \\[6pt] (e^2 - 1)e^{-2z}, & z \geq 1 \end{cases}}$$

**Проверка нормировки:**

$$\int_0^1 (1 - e^{-2z}) \, dz + \int_1^{\infty} (e^2 - 1)e^{-2z} \, dz$$
$$= \left[z + \frac{e^{-2z}}{2}\right]_0^1 + (e^2 - 1) \left[-\frac{e^{-2z}}{2}\right]_1^{\infty}$$
$$= \left(1 + \frac{e^{-2}}{2} - 0 - \frac{1}{2}\right) + (e^2 - 1) \cdot \frac{e^{-2}}{2}$$
$$= 1 + \frac{e^{-2} - 1}{2} + \frac{e^2 - 1}{2e^2} = 1 + \frac{e^{-2} - 1}{2} + \frac{1 - e^{-2}}{2} = 1$$
